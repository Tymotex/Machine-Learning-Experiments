Quick set-up and some example plots:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
np.seterr(divide='ignore', invalid='ignore')  # Prevents the division by zero warning when plotting
```

```python
x = np.linspace(0, 6, 100)
y = 2 * (x **3 ) * 1 / (np.exp(x) - 1)
plt.plot(x, y)
```

Adding axis labels:

```python
plt.plot(x, y, "ko-")  # Third argument specifies the linestyle: k for black, o for symbol, - for solid line
                       # Other colours are b for blue, r for red, g for green, ...
plt.xlabel("Wavelength (micrometers)")
plt.ylabel("Spectral Radiance")
plt.title("Not a blackbody radiation graph")
```
